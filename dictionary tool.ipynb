```python
import json
import nltk
filename = 'dictionary.txt'

words = {}
i=0
with open(filename) as fh:
    for line in fh:
        word= line.rstrip().split('  ', 1)
        if len(word)>=2:
            des=str(word).split(',',1)
            word=des[0]
            word=word[2:-1]
            word=word.lower()
            try:
                if word[len(word)-1]=='-':
                    word=word[:len(word)-2]
            except:
                i=i
            des=des[1]
            des=des[2:-2]
            des=des.lower()
            words[word] = des
        i=i+1
    
#print(words)
with open("dictionary1.json", "w") as write_file:
    json.dump(words,(write_file), ensure_ascii=False)
```
